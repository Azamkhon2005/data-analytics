```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers_data = pd.read_csv(r'olist_customers_dataset.csv')
geolocation_data = pd.read_csv(r'olist_geolocation_dataset.csv')
order_items_data = pd.read_csv(r'olist_order_items_dataset.csv')
order_payments_data = pd.read_csv(r'olist_order_payments_dataset.csv')
order_reviews_data =pd.read_csv(r'olist_order_reviews_dataset.csv')
orders_data = pd.read_csv(r'olist_orders_dataset.csv')
products_data = pd.read_csv(r'olist_products_dataset.csv')
sellers_data = pd.read_csv(r'olist_sellers_dataset.csv')
product_category_name_translation = pd.read_csv(r'product_category_name_translation.csv')
```

```python
df = orders_data.merge(order_items_data,how='inner', on='order_id').merge(customers_data,how='inner',on='customer_id')
```

```python
df.merge(order_payments_data[['order_id','payment_type','payment_value']],how='inner',on='order_id')
```

```python
df['purchase_price'] = df['order_item_id']*df['price']
```

```python
df = df[df['order_status']=='delivered']
```

```python
df = df.merge(geolocation_data[['geolocation_zip_code_prefix','geolocation_lat','geolocation_lng']],how='inner',left_on='customer_zip_code_prefix',right_on='geolocation_zip_code_prefix')
```

```python
product_category = products_data.merge(product_category_name_translation[['product_category_name','product_category_name_english']],how='inner',on='product_category_name')[['product_id','product_category_name_english']]
product_category.rename({'product_category_name_english':'product_category'},axis=1,inplace=True)
```

```python
df = df.merge(product_category,how='inner',on='product_id')
```

```python
df = df.drop(['shipping_limit_date','order_approved_at','order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date','order_item_id','price','order_status'],axis=1)
```

```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
```

```python
df.rename({'order_purchase_timestamp':'order_purchase','geolocation_lat':'latitude','geolocation_lng':'longtitude'},axis=1,inplace=True)
```

```python
df.drop(['customer_zip_code_prefix','geolocation_zip_code_prefix'],axis=1,inplace=True)
```

```python
df.to_csv('cleaned_data.csv',sep=',', index=False)
```
